```python
from sklearn.linear_model import Perceptron
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X = np.array([[0,0],[0,1],[1,0],[1,1]])
# y = np.array([0,1,1,1])
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])
```

```python
# 산포도
plt.scatter(X[:,0], X[:,1], c = y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
model = Perceptron(verbose=1)

model.fit(X,y)
```

```python
print(model.coef_)
print(model.intercept_)
print(model.n_iter_)
```

```python
plt.scatter(X[:, 0], X[:,1], c=y)
plt.xlabel('x1')
plt.ylabel('x2')

w1 = model.coef_[0,0]
w2 = model.coef_[0,1]
b = float(model.intercept_)

x_intercept = - b/w1
y_intercept = - b/w2

plt.plot([0, y_intercept], [x_intercept,0] , c= 'red')
plt.show()
```

```python
X_test = X
pred = model.predict(X_test)
print(pred)
```
