```python
import numpy as np

np.random.seed(85)
x = np.arange(0, 10)
y1 = x * 2
np.corrcoef(x, y1)

```

```python
import numpy as np

x = np.arange(0, 10)
y2 = x ** 3
np.corrcoef(x, y2)

```

```python
import numpy as np

x = np.arange(0, 10)
y3 = np.random.randint(0,100,size=10)
np.corrcoef(x, y3)

```

```python
result = np.corrcoef((x,y2,y3))
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(result)
plt.colorbar
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'x':x,'y2':y2,'y3':y3})
sns.pairplot(df)
```

```python
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
tips = sns.load_dataset("tips")

sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
tips.head(7)
```

```python
tips.shape
```

```python
sns.distplot(tips['tip'],kde=False,bins=10)
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
sns.distplot(tips['tip_pct'],kde=True,bins=20)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='smoker', data=tips)
```

```python
sns.relplot(data=tips, x = "total_bill", y="tip", col="time",hue = "smoker", style="smoker", size="size")
```

```python
ax = sns.regplot(data=tips, x="total_bill", y = 'tip')
ax.set_xlabel('Total bill')
ax.set_ylabel('Tip')
ax.set_title('Total Bill and Tip')
```

```python
sns.pairplot(tips)
```
