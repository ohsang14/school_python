```python
import csv

f = open('weather.csv', encoding='utf-8')  # CSV 파일 열기 => 파일 핸들 f에 저장
data = csv.reader(f)  # reader 생성자 => csv 파일 읽기
header = next(data)   # 헤더를 제거

max_wind = 0.0  # max_wind 초기값
for row in data:    # 반복 루프
    if row[3] == '': # 평균  풍속 데이터가 없는 경우 0을 처리
        wind = 0
    else:
        wind = float(row[3])  # 평균 풍속 데이터(str)를 실수(float)로 변환
        if max_wind < wind:   # 현재까지의 최대 풍속보다 크면 최대값 업데이트
            max_wind = wind

print('최대 풍속 =', max_wind, 'm/s')
f.close()
```

```python
import csv

f = open('weather.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data)

min_hot = float('inf')
max_wind = 0.0
total_wind = 0.0
count_wind = 0

for row in data:
    if row[1] == '':
        hot = float('inf')
    else:
        hot = float(row[1])
        if min_hot > hot:
            min_hot = hot

    if row[3] == '':
        wind = 0
    else:
        wind = float(row[3])
        if max_wind < wind:
            max_wind = wind

    if row[3] != '':
        total_wind += wind
        count_wind += 1

if count_wind > 0:
    avg_wind = total_wind / count_wind
else:
    avg_wind = 0

print('최대 풍속 =', max_wind, 'm/s')
print('평균 기온 최소값 =', min_hot, '도')
print('평균 풍속 =', avg_wind, 'm/s')

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('weather.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data)

monthly_wind = [0 for x in range(12)]
days_counted = [0 for x in range(12)]

for row in data:
    month = int(row[0][5:7])
    if row[3] != '':
        wind = float(row[3])
        monthly_wind[month - 1] += wind
        days_counted[month - 1] += 1

for i in range(12):
    monthly_wind[i] /= days_counted[i]

plt.plot(monthly_wind, 'blue')
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('weather.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data)

monthly_wind = [0 for x in range(12)]
monthly_max_wind = [0 for x in range(12)]
days_counted = [0 for x in range(12)]

for row in data:
    month = int(row[0][5:7]) - 1
    if row[3] != '':
        wind = float(row[3])
        monthly_wind[month] += wind
        days_counted[month] += 1

        if wind > monthly_max_wind[month]:
            monthly_max_wind[month] = wind

for i in range(12):
    monthly_wind[i] /= days_counted[i]

plt.plot(monthly_wind, 'blue', label='평균 풍속')
plt.plot(monthly_max_wind, 'red', label='최대 풍속')
plt.legend()
plt.show()
f.close()
```
